### Data Pre-processing 

```python
# Checking the null values

movies.isnull().sum()
```

```python
# Dropping the null values 

movies.dropna(inplace = True)
```

```python
movies.isnull().sum()
```

```python
# Checking the duplicte rows

movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
# Taking out all the genres from the dictionary by using function

def convert(genres):
    import ast 
    l = []
    for i in ast.literal_eval(genres):
        l.append(i['name'])
    return l
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
# Creating a function to take out the actors and actress name 

def convert2 (keywords) :
    import ast
    l = []
    count = 0
    for i in ast.literal_eval(keywords):
        if count != 3 :
            l.append(i['name'])
            count +=1
        else :
            break 
    return l 
    
```

```python
movies['cast'] = movies['cast'].apply(convert2)
```

```python
movies.head()
```

```python
# Creating a function to take out the director name 

def convert3 (crew) :
    import ast
    l = []
    for i in ast.literal_eval(crew):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
    return l 
    
```

```python
movies['crew'] = movies['crew'].apply(convert3)
```

```python
movies.head(1)
```

```python
# Converting overview columns values into list 

movies['overview'] = movies['overview'].apply(lambda x : x.split())
```

```python
# Removing space between two words 

movies['genres'] = movies['genres'].apply(lambda x : [i.replace(' ','') for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x : [i.replace(' ','') for i in x])
movies['cast'] = movies['cast'].apply(lambda x : [i.replace(' ','') for i in x])
movies['crew'] = movies['crew'].apply(lambda x : [i.replace(' ','') for i in x])
```

```python
movies.head(1)
```

```python
movies ['tags'] = movies['overview']+ movies['genres'] + movies['keywords']+ movies['cast'] + movies['crew'] 
```

```python
movies.head(1)
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df
```

```python
# Jioning list elements into a string with space 

new_df['tags'] =  new_df['tags'].apply(lambda x : " ".join(x))
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x : x.lower())
```

```python
new_df.head()
```

```python
new_df['tags'][0]
```

### Verctorizing the text 

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features = 5000,stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

### Applying steming

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
#Creating a steming function

def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

### Calculating the cosine distance

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

### Creating a recommender function 

```python
def recommender(movie):
    # for finding the index of a movie
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list =  sorted(list(enumerate(distances)),reverse = True, key = lambda x:x[1])[1:6]
    
    for i in movies_list :
        print(new_df.iloc[i[0]].title)
```

```python
# finding the movies index and sorting them in descending order

sorted(list(enumerate(similarity[0])),reverse = True, key = lambda x:x[1])[1:6]
```

```python
recommender('Batman Begins')
```
